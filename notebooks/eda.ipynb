```python
plt.figure(figsize=(20,10))
plt.title('Relationship plot', fontsize=(20),color='red')
## first plot
plt.subplot(2,3,1)
sns.scatterplot(x=df['sepal_length'],y=df['sepal_width'],hue=df['species'])


## second plot
plt.subplot(2,3,2)
sns.scatterplot(x=df['sepal_length'],y=df['petal_length'],hue=df['species'])

## third plot
plt.subplot(2,3,3)
sns.scatterplot(x=df['sepal_length'],y=df['petal_width'],hue=df['species'])


## fourth plot
plt.subplot(2,3,4)
sns.scatterplot(x=df['sepal_width'],y=df['petal_length'],hue=df['species'])


## fifth plot
plt.subplot(2,3,5)
sns.scatterplot(x=df['sepal_width'],y=df['petal_width'],hue=df['species'])


## six plot
plt.subplot(2,3,6)
sns.scatterplot(x=df['petal_length'],y=df['petal_width'],hue=df['species'])
plt.show()
```

```python
columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors=['red', 'green', 'blue','yellow']
```

```python
plt.figure(figsize=(20,10))
j=1
sns.set_theme(style="darkgrid")
for i in columns:
    plt.subplot(2,2,j)
    plt.title(i)
    sns.histplot(df[i],color=colors[j-1],kde=True)
    j=j+1
plt.show()
```

```python
plt.figure(figsize=(20,10))
j=1
sns.set_theme(style="darkgrid")
for i in columns:
    plt.subplot(2,2,j)
    plt.title(i)
    sns.boxplot(x=df['species'],y=df[i],color=colors[j-1])
    j=j+1
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
new_df=pd.DataFrame(scaler.fit_transform((df[columns])),columns=columns)
```

```python
plt.figure(figsize=(20,10))
j=1
sns.set_theme(style="darkgrid")
for i in columns:
    plt.subplot(2,2,j)
    plt.title(i)
    # plt.axis('off')
    sns.boxplot(x=df['species'],y=new_df[i],color=colors[j-1],native_scale=True)
    j=j+1
plt.show()
```
